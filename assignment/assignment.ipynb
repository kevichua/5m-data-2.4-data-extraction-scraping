### Question 1

The scraping of `https://www.scrapethissite.com/pages/forms/` in the last section assumes a hardcoded (fixed) no of pages. Can you improve the code by removing the hardcoded no of pages and instead use the `»` button to determine if there are more pages to scrape? Hint: Use a `while` loop.

```python
import requests
import time
from bs4 import BeautifulSoup

def parse_and_extract_rows(soup: BeautifulSoup):
    """
    Extract table rows from the parsed HTML.

    Args:
        soup: The parsed HTML.

    Returns:
        An iterator of dictionaries with the data from the current page.
    """
    header = soup.find('tr')
    headers = [th.text.strip() for th in header.find_all('th')]
    teams = soup.find_all('tr', 'team')
    for team in teams:
        row_dict = {}
        for header, col in zip(headers, team.find_all('td')):
            row_dict[header] = col.text.strip()
        yield row_dict
        
```

Answer:

```python
# Enter your solution here
all_rows = []
page = 1

while True:
    # 1. Request the current page
    response = requests.get(f"https://www.scrapethissite.com/pages/forms/?page_num={page}")
    soup = BeautifulSoup(response.text, "html.parser")

    # 2. Extract rows from this page
    rows = list(parse_and_extract_rows(soup))
    all_rows.extend(rows)

    # 3. Check if the >> (next) button exists
    # This part is troublesome, need to inspect the source to find the >> button 
    # ctrl + u, then ctrl + f, then type in to find aria-label="Next"
    
    # .find function in BeautifulSoup
    # First argument: the HTML tag to search for → <a>
    # Second argument: a filter dictionary → only match <a> tags WHERE aria-label="Next"
    # Returns the FIRST matching element found, or None if not found
    next_button = soup.find("a", {"aria-label": "Next"})

    if next_button is None:
        print(f"Reached the last page at page {page}")
        break  # break out of loop

    # 4. Move to the next page
    page += 1
    time.sleep(1)
```

```python
# print the number of rows extracted
print(f"Total rows scraped: {len(all_rows)}")
```
